```python
from transformers import pipeline

gen = pipeline("text-generation")
```

```python
prompt = "Once upon a time there was a"
```

```python
predict = gen(prompt, max_length = 20)
```

```python
print(predict[0]["generated_text"])
```

```python
predict = gen(prompt, max_length = 25)
```

```python
print(predict[0]["generated_text"])
```
